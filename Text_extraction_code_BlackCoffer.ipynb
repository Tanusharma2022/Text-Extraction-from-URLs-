```python
#loading the file
df=pd.read_excel('Input.xlsx') #file name should match case
print(df.head())
```

```python
answers= requests.get("https://insights.blackcoffer.com/ai-and-ml-based-youtube-analytics-and-content-creation-tool-for-optimizing-subscriber-engagement-and-content-strategy/")
answers.raise_for_status()
extract = BeautifulSoup(answers.content, 'html.parser')
print(extract.prettify())

```

```python
def extract_text_from_articles(url):
    try:
        answers= requests.get(url)
        answers.raise_for_status()  # Check for errors if they arise while requesting
        extract = BeautifulSoup(answers.content, 'html.parser')

        # Extracting the Topic of the articles in input.xlx file
        topics = extract.find('h1').get_text(strip=True) if extract.find('h1') else 'No heading Found'
        #Extract the article text
        #looking for the class containing content; I found the span tag with class "td-post-content" contain the required text
        A_body = extract.find(class_= 'td-post-content')
        if A_body:
            B_text = A_body.get_text(separator='\n', strip=True)
        else:
            B_text = 'No Article Body Found'

        return topics, B_text
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return None, None  #error handlin

# Creating a filename based on URL_ID
def extracted_text_to_file(url_id, topics, all_texts):
    filename = f"{url_id}.txt"
    with open(filename, 'w', encoding='utf-8') as file:
        file.write(f"{topics}\n\n")
        file.write(all_texts)

```

```python
def main():
    # DataFrame has columns 'URL_ID' and 'URL'
    for index, row in df.iterrows():
        url_id = row['URL_ID']
        url = row['URL']

        topics, B_text = extract_text_from_articles(url)
        if topics and B_text:
            extracted_text_to_file(url_id, topics, B_text)
            print(f"Article of {url_id}")
        else:
            print(f"Article not saved {url_id}")
main()
```
